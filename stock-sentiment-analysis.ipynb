```python
df = pd.read_csv('/kaggle/input/stock-sentiment-analysis/Stock_Dataa.csv',encoding='iso-8859-1')
df.head()
```

```python
len(df)
```

```python
train = df[df['Date']<'20150101']
test = df[df['Date']>'20141231']
print(len(train))
print(len(test))
```

```python
data = train.iloc[:,2:27]
data.replace('[^a-zA-Z]',' ',regex=True,inplace=True)
```

```python
data.head()
```

```python
data.columns
```

```python
new_list = [i for i in range(25)]
new_Index = [str(i) for i in new_list]
data.columns = new_Index
```

```python
data.head()
```

```python
for index in new_Index:
    data[index] = data[index].str.lower()   
```

```python
#combining all the 25 columns of the data
' '.join(str(x) for x in data.iloc[0,:25])
```

```python
#putting into a list of all rows with cmbined 25 columns
headlines = []
for i in range(len(data)):
    headlines.append(' '.join(str(x) for x in data.iloc[i,:25]))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
countvector = CountVectorizer(ngram_range=(2,2))
traindataset = countvector.fit_transform(headlines)
```

```python
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
test_transform = []
for i in range(len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[i,2:27]))
```

```python
test_dataset = countvector.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

```python
print(predictions)
```

```python
#Importing libraries to test accuracy
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(classification_report(test['Label'],predictions))
```

```python
print(confusion_matrix(test['Label'],predictions))
```

```python
print(accuracy_score(test['Label'],predictions))
```
